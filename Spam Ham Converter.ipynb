```python
import pandas as pd
import numpy as np
data = pd.read_csv('sms-spam-collection-dataset/spam.csv', encoding='latin-1')
```

```python
data = data[['v1']]
lst = []
for i in data.values:
    lst.extend(i)
lst
```

```python
def spam_ham_converter(lst):
    new_a = np.array([])
    for i in lst:
        if i == 'ham':
            new_a = np.append(new_a, 0)
            continue
        new_a = np.append(new_a, 1)
    return new_a
            
y = spam_ham_converter(lst)
y
```
